Question 7: Which genre has the highest energy score?
    
By sorting by energy scoring, the highest energy score is 98, which goes to two songs:
    Pom Poms, by Jonas Brothers, which is in the genre boy band,
    and Hello, by Martin Solveig which is in the genre big room. 

A second interpretation of the question is as follows:
I made each genre a category and found the mean of the energy score for each genre. Then, sorted by energy score (see analysis below). 
So, the genre with the highest mean energy score as a combined category is:
French Indie Pop (energy score = 89). 

```python
grouped = all_top_songs.groupby('top genre')
avgs = grouped.mean()
avgs
```

```python
nrgy = avgs.sort_values(
...     by="nrgy",
...     ascending=False
... )
nrgy.head()
```
